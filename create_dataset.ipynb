```python
# validationからランダムでtestにして保存
num_test_videos = 228
test_ids = []
c = 0
while c < num_test_videos:
    random_recipe_type = random.choices(list(subsets['validation'].keys()))
    random_recipe_ID = random.choices(list(subsets['validation'][random_recipe_type[0]].keys()))
    if random_recipe_ID[0] in test_ids:
        continue
    test_ids.append(random_recipe_ID[0])
    c += 1
with open("test_ids",'w') as f:
    for i in test_ids:
        f.write(i+"\n")
#print(test_ids)
l_duplicate = [x for x in set(test_ids) if test_ids.count(x) > 1]

#if '84i8Qdnyd0k' l_duplicate
```

```python
train_image_paths = []
train_image_captions = []
val_image_paths = []
val_image_captions = []
test_image_paths = []
test_image_captions = []
word_freq = Counter()

for subset, value in subsets.items():
    for recipe_type, value2 in value.items():
        for name,value3 in value2.items():
            for id,sentence in value3.items():

                id = str(id).zfill(2)
                image_name = id + '-000.jpg'
                image_path = os.path.join(image_dir,subset,recipe_type,name,image_name)
                if not os.path.exists(image_path):
                    continue
                im_sentence = sentence
                tokens = [i.lower() for i in im_sentence.split()]
                word_freq.update(tokens)
                if subset == 'training':
                    train_image_paths.append(image_path)
                    train_image_captions.append([tokens])
                elif subset == 'validation':
                    if name in test_ids:
                        test_image_paths.append(image_path)
                        test_image_captions.append([tokens])                        
                    else:    
                        val_image_paths.append(image_path)
                        val_image_captions.append([tokens])
```

```python
min_word_freq = 5
captions_per_image = 1
max_len = 50
print("dd")
# create word map
words = [w for w in word_freq.keys() if word_freq[w] > min_word_freq]
word_map = {k: v + 1 for v, k in enumerate(words)}
word_map['<unk>'] = len(word_map) + 1
word_map['<start>'] = len(word_map) + 1
word_map['<end>'] = len(word_map) + 1
word_map['<pad>'] = 0
print("dd")
#
base_filename = 'coco' + '_' + "5" + '_cap_per_img_' + str(min_word_freq) + '_min_word_freq'
print("dd")
# Save word map to a JSON
with open(os.path.join('dataset', 'WORDMAP_' + base_filename + '.json'), 'w') as j:
    json.dump(word_map, j)

```

```python
output_folder = "dataset"
seed(123)
for impaths, imcaps, split in [(train_image_paths, train_image_captions, 'TRAIN'),
                                   (val_image_paths, val_image_captions, 'VAL'),
                                   (test_image_paths, test_image_captions, 'TEST')]:
    #print(len(imcaps[i]))
    #break
    with h5py.File(os.path.join(output_folder, split + '_IMAGES_' + base_filename + '.hdf5'), 'a') as h:
        # Make a note of the number of captions we are sampling per image
        h.attrs['captions_per_image'] = captions_per_image
        # Create dataset inside HDF5 file to store images
        images = h.create_dataset('images', (len(impaths), 3, 256, 256), dtype='uint8')
        print("\nReading %s images and captions, storing to file...\n" % split)
        enc_captions = []
        caplens = []
        
        for i, path in enumerate(tqdm(impaths)):

            # Sample captions

            if len(imcaps[i]) < captions_per_image:
                captions = imcaps[i] + [choice(imcaps[i]) for _ in range(captions_per_image - len(imcaps[i]))]
                #print(captions)
                #break
            else:
                captions = sample(imcaps[i], k=captions_per_image)

            # Sanity check
            assert len(captions) == captions_per_image

            # Read images
            img = imread(impaths[i])
            if len(img.shape) == 2:
                img = img[:, :, np.newaxis]
                img = np.concatenate([img, img, img], axis=2)
            img = imresize(img, (256, 256))
            img = img.transpose(2, 0, 1)
            assert img.shape == (3, 256, 256)
            assert np.max(img) <= 255
            
            # Save image to HDF5 file
            images[i] = img
            for j, c in enumerate(captions):
                # Encode captions
                enc_c = [word_map['<start>']] + [word_map.get(word, word_map['<unk>']) for word in c] + [
                    word_map['<end>']] + [word_map['<pad>']] * (max_len - len(c))
                #print(c)
                #print(enc_c)
            
                # Find caption lengths
                c_len = len(c) + 2

                enc_captions.append(enc_c)
                caplens.append(c_len)
        
        # Sanity check
        assert images.shape[0] * captions_per_image == len(enc_captions) == len(caplens)

        # Save encoded captions and their lengths to JSON files
        with open(os.path.join(output_folder, split + '_CAPTIONS_' + base_filename + '.json'), 'w') as j:
            json.dump(enc_captions, j)

        with open(os.path.join(output_folder, split + '_CAPLENS_' + base_filename + '.json'), 'w') as j:
            json.dump(caplens, j)
            
            
```

```python
#doc2vecの学習
trainings = []
c = 0
for subset, value in subsets.items():
    for recipe_type, value2 in value.items():
        for name,value3 in value2.items():
            for id,sentence in value3.items():
                

                
                im_sentence = gensim.utils.simple_preprocess(sentence)
                trainings.append(TaggedDocument(im_sentence,str(c)))
                c += 1
m = Doc2Vec(documents= trainings, dm = 1, size=300, window=8, min_count=10, workers=4)
```
